<h1>Indexes :</h1>

```python
people = {
    "first" : ["Corey", "Jane", "John"],
    "last" : ["Schafer", "Doe", "Doe"],
    "email" : ["CoreyMSchafer@gmail.com", "JaneDoe@gmai.com", "JohnDoe@gmail.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.set_index('email',inplace = True)
```

```python
df
```

```python
df.index
```

```python
df.loc['CoreyMSchafer@gmail.com']
```

```python
df.loc['CoreyMSchafer@gmail.com','last']
```
