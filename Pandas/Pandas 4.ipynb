<h1>Filtering  :</h1>

```python
people = {
    "first" : ["Corey", "Jane", "John"],
    "last" : ["Schafer", "Doe", "Doe"],
    "email" : ["CoreyMSchafer@gmail.com", "JaneDoe@gmai.com", "JohnDoe@gmail.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['last'] == 'Doe'
```

```python
filt = (df['last'] == 'Doe')
```

```python
df[filt]   # it return all the row that have last name as Doe
```

```python
df.loc[filt]
```

```python
df.loc[filt,['first', 'email']]
```

```python
filt = (df['last'] == 'Doe') & (df['first'] == 'John')
df.loc[filt,'email']
```

```python
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')
df.loc[filt, 'email']
```

```python
df.loc[~filt, 'email']
```
